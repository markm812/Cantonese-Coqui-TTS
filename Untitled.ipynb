```python
import os
from pydub import AudioSegment
import shutil

arr = os.listdir('./zh-HK/clips/')
```

```python
arr
```

```python
for f in range(len(arr)):
    # progress
    print(f"Progress: {f+1}/{len(arr)+1}")

    # files                                                                         
    src = os.path.join("zh-HK","clips",arr[f])
    dst = os.path.join("zh-HK","clips",arr[f].split(".")[0] + ".wav") 

    # convert wav to mp3
    print(src)
    sound = AudioSegment.from_file(src,"mp3")
    sound.export(dst, format="wav")

    # remove mp3 file
    os.remove(src)
```
